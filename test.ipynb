```python
import numpy as np
from Zebrafish_model import *
import matplotlib.pyplot as plt
from PIL import Image
plt.style.use('seaborn-whitegrid')
```

```python
# Experiment 3
p3 = np.load('results/exp3/p_val.npy')
s3 = np.load('results/exp3/s_val.npy')
g3 = np.load('results/exp3/g_val.npy')
pe3 = np.load('results/exp3/pe_val.npy')
se3 = np.load('results/exp3/se_val.npy')
ge3 = np.load('results/exp3/ge_val.npy')
```

```python
def update_array(x):
  temp = x[0]
  for i in range(1, len(x)):
    temp1 = x[i]
    x[i] = x[i] - temp
    temp = temp1
  return x
```

```python
s3 = update_array(s3)
g3 = update_array(g3)
pe3 = update_array(pe3)
```

```python
g_perc = (sum(g3)/sum(p3))*100
s_perc = (sum(s3)/sum(p3))*100
pe_perc = (sum(pe3)/sum(p3))*100
```

```python
print(f'Glucurodination: {g_perc}%, Sulfation: {s_perc}%, Paracetamol Excretion: {pe_perc}%')
```
